### Basic

```python
"""
Note:-
- In any widget , 1st argument is always "root"
"""

from tkinter import *

root = Tk()
# root.geometry("400x400")

#Title
root.title("My New App")

#Function
def change_lable(event = None):
    lbl.config(text="hi")

# Label
lbl = Label(root, text="Welcome to the app")
lbl.grid(row=0 , column= 0 )

# Button 1
btn = Button(root, text="Click me", command=change_lable)
btn.grid(row=1, column=0)

# Button 2
btn2 = Button(root, text= "close window", command= root.destroy)
btn2.grid(row=2, column=0)

#CheckButton -> Checkbox
chck_btn = Checkbutton(root)
chck_btn.grid(row= 4, column=0)

# Single line input field -> Use the Entry()
entry1 = Entry(root)
entry1.grid(row= 3, column=0)

# Multi-line input field -> Use the Text()
entry2 = Text(root, background= "#1f1f1f", font="Helvetica", fg= "White", insertbackground="white")
entry2.grid(row= 5, column=0)


root.mainloop()
```

### Calculator

```python
from tkinter import *

def click(event):
    global expression
    text = event.widget.cget("text")
    if text == "=":
        if expression:                  # If the expression is not empty
            result = eval(expression)   # Evaluates the expression using eval() 
            expression = str(result)    # 
            input_var.set(expression)

    elif text == "C":
        expression = ""
        input_var.set("")
    else:
        expression += text
        input_var.set(expression)

root = Tk()
root.title("Simple Calculator")
expression =  ""       # For storing the expression

input_var = StringVar()      # if did like " input_var = "" ", then Without StringVar, you'd have to manually update the input field every time the value changes.
entry = Entry(root, text = input_var, font = ("Arial", 18),justify="right")
entry.grid(row=0, column=0,columnspan=4, ipadx=10, ipady=10)

buttons = [
    "7", "8", "9", "/",
    "4", "5", "6", "*",
    "1", "2", "3", "-",
    "C", "0", "=", "+"
]

btn_row = 1
btn_column = 0

for button in buttons:
    btn = Button(root, text=button,font = ("Arial", 10), padx=20, pady=20)
    btn.grid(row=btn_row, column= btn_column)
    btn.bind("<Button-1>", click)    #the event triggered when the left mouse button is pressed down.
    btn_column +=1

    if btn_column>3:
        btn_column = 0
        btn_row +=1

root.mainloop()
```

### Threading

```python
#===================================
# Without Threading

# Import Module 
from tkinter import *
import time 
from threading import *

 # Create Object 
root = Tk() 
# Set geometry 
root.geometry("400x400") 

def work(): 
    print("sleep time start") 

    for i in range(10): 
        print(i) 
        time.sleep(1) 

    print("sleep time stop") 

# Create Button 
Button(root, text="Click Me",command=work).pack() 

# Execute Tkinter 
root.mainloop()

#==========================
#With Threading

# Import Module 
from tkinter import *
import time 
from threading import *

 # Create Object 
root = Tk() 

# Set geometry 
root.geometry("400x400") 

# use threading 
def threading(): 
    # Call work function 
    t1=Thread(target=work) 
    t1.start() 

# work function 
def work(): 
    print("sleep time start") 
    for i in range(10): 
        print(i) 
        time.sleep(1) 
    print("sleep time stop") 

# Create Button 
Button(root,text="Click Me",command = threading).pack() 

# Execute Tkinter 
root.mainloop()
```

### Top Level Widget
- Used to open one window over another

```python
from tkinter import *


root = Tk()
root.geometry("200x300")
root.title("main")
l = Label(root, text = "This is root window")


top = Toplevel()
top.geometry("180x100")
top.title("toplevel")
l2 = Label(top, text = "This is toplevel window")

l.pack()
l2.pack()

root.mainloop()
```

### Creating Multiple toplevels over one another 

```python
from tkinter import *


# Create the root window
# with specified size and title
root = Tk() 
root.title("Root Window") 
root.geometry("450x300") 

# Create label for root window
label1 = Label(root, text = "This is the root window")

# define a function for 2nd toplevel 
# window which is not associated with 
# any parent window
def open_Toplevel2(): 
	
	# Create widget
	top2 = Toplevel() 
	
	# define title for window
	top2.title("Toplevel2")
	
	# specify size
	top2.geometry("200x100")
	
	# Create label
	label = Label(top2,
				text = "This is a Toplevel2 window")
	
	# Create exit button.
	button = Button(top2, text = "Exit", 
					command = top2.destroy)
	
	label.pack()
	button.pack()
	
	# Display until closed manually.
	top2.mainloop()
	
# define a function for 1st toplevel
# which is associated with root window.
def open_Toplevel1(): 
	
	# Create widget
	top1 = Toplevel(root)
	
	# Define title for window
	top1.title("Toplevel1")
	
	# specify size
	top1.geometry("200x200")
	
	# Create label
	label = Label(top1,
				text = "This is a Toplevel1 window")
	
	# Create Exit button
	button1 = Button(top1, text = "Exit",
					command = top1.destroy)
	
	# create button to open toplevel2
	button2 = Button(top1, text = "open toplevel2",
					command = open_Toplevel2)
	
	label.pack()
	button2.pack()
	button1.pack()
	
	# Display until closed manually
	top1.mainloop()

# Create button to open toplevel1
button = Button(root, text = "open toplevel1",
				command = open_Toplevel1)
label1.pack()

# position the button
button.place(x = 155, y = 50)

# Display until closed manually
root.mainloop()
```
