### Pie chart basics
- Use pie charts if you want to compare atmost 5 items

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices = [60, 40, 30 , 20]

#Creating labels
labels = ["sixty", "forty", "Extra_1" , "Extra_2"]

#Creating Colors
colors = ["#008fd5", "#fc4f30", "#e5ae37", "#6d904f"]

#---------------------
#Plotting
plt.pie(slices, labels= labels, colors=colors, wedgeprops={"edgecolor" : "black"})

#---------------------------------
# 3. FORMATTING

#Adding title to the plot
plt.title("My Pie chart")                 

#It will compress the graph into smaller size
plt.tight_layout()

#-------------------
plt.show()        
```

### Advance Pie Chart

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

#Used to give special imphasise on certain pie of the chart. "0" means default , "0.1" means 10% of it exploded out fom the radius
explode = [0, 0, 0, 0.1, 0]
#---------------------
"""
- startangle  -> To rotate a chart
- autopct     -> Add % to each pie
- shadow      -> Addes shadow to the pie chart 
- wedgeprops  -> Add color to the boundary of two adjacent pie
"""
#Plotting
plt.pie(slices, labels= labels, explode=explode, shadow= True, startangle= 90, autopct= "%1.1f%%" ,wedgeprops={"edgecolor" : "black"})

#---------------------------------
# 3. FORMATTING

#Adding title to the plot
plt.title("My Pie chart")                 

#It will compress the graph into smaller size
plt.tight_layout()

#-------------------
plt.show()        
```
